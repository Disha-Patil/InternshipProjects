# Relationsip Analysis

```python
corelation = StudentsPerformance.corr()
```

```python
# Heatmap 

sns.heatmap(df.corr())
```

```python
# PairPlot

sns.pairplot(StudentsPerformance)
```

```python
# Relation Plot

sns.relplot(x= 'math score', y='writing score', hue='gender', data= StudentsPerformance)
sns.relplot(x= 'math score', y='reading score', hue='gender', data= StudentsPerformance)
```

In above Relation Plots we can observe that greater number of Female students are obtaining higher marks in Writing and Reading, whereas greater number of Male students are obtaining higher marks in Mathematics.
