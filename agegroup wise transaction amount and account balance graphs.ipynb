```python
#importing pandas
import pandas as pd
```

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#reading the csv file bank transactions
df=pd.read_csv('bank_transactions.csv')
```

```python
#importing datetime
from datetime import datetime
```

```python
#retrieving today's date
x=datetime.today()
```

```python
#declaring empty list
age=[]
```

```python
#convetring the column CustomerDOB into datetime form
df.CustomerDOB = pd.to_datetime(df.CustomerDOB)
```

```python
#extracting year of date of birth
year=df.CustomerDOB.dt.year
```

```python
#extracting month of date of birth
month=df.CustomerDOB.dt.month
```

```python
#extracting year of today's date
y=int(x.strftime('%Y'))
```

```python
#extracting month of today's date
m=int(x.strftime('%m'))
```

```python
#extracting the age of customers and adding them into the empty list declared above
for i in range(0,len(year)):
    if(month[i]>m and (month[i]-m)<6):
        a=y-year[i]
    elif(month[i]<m and (m-month[i])<6):
        a=y-year[i]
    elif(month[i]<m and (m-month[i])>=6):
        a=y-year[i]+1
    else:
        a=y-year[i]+1
    age.append(a)
```

```python
#calling the list
age
```

```python
#finding average of all ages which will replace the negative ages as well as nan
sum=0
for i in range(0, len(age)):
    if(age[i]>0):
        sum=sum+age[i]
avg=sum/len(age)
d=int(avg)
for i in range(0, len(age)):
    if(age[i]<0):
        age[i]=d
    if(str(age[i]).upper()=='NAN'):
        age[i]=int(str(d))
```

```python
#testing if the replacement was successful
age[222]
```

```python
#adding a column age into the table
df['age']=age
```

```python
#calling the changed table
df
```

```python
#renaming the column of the table for conveniance
df=df.rename(columns={'TransactionAmount (INR)': 'TransactionAmount'})
```

```python
#calling the changed table
df
```

```python
#finding the agegroup of customers
agegroup=[]
for i in range(0, len(df.age)):
    if(int(age[i]/10)==0):
        agegroup.append('0-10')
    else:
        agegroup.append(str(int(age[i]/10)*10+1)+'-'+str(int(age[i]/10)*10+10))
```

```python
#calling the agegroup
agegroup
```

```python
#mqking a new column agegroup in the table
df['agegroup']=agegroup
```

```python
#testing if the agegroup shown is correct
agegroup[222]
```

```python
#extracting unique elements of agegroup and making a new list agegroup1
agegroup1=list(set(agegroup))
```

```python
#calling the list agegroup1
agegroup1
```

```python
#declaring empty list and adding zeros in it whose numbers are equal to the number of unique elements of the previous list 
m=[]
for i in agegroup1:
    m.append(0)
```

```python
#finding the sum of customer account balance of each agegroup of all locations
for i in range(0, len(agegroup1)):
    for j in range(0, len(df.agegroup)):
        if(agegroup1[i].upper()==str(df.agegroup[j]).upper()):
            if(str(df.CustAccountBalance[j]).upper()=='NAN'):
                m[i]=m[i]+0
            else:
                m[i]=m[i]+float(df.CustAccountBalance[j])
```

```python
#representing the customer account balance of all the agegroups of all locations
m
```

```python
#plotting the pie chart
plt.pie(m, labels=agegroup1)
plt.show()
```

```python
#plotting the bar graph
plt.bar(agegroup1, m)
plt.xticks(rotation=45)
plt.show()
```

```python
#plotting the line graph
plt.plot(agegroup1, m)
plt.xticks(rotation=45)
plt.show()
```
